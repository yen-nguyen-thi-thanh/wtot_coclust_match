```python
### import packages
import ot
import numpy as np
import pandas as pd
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.ticker as mtick
from mpl_toolkits.axes_grid1 import make_axes_locatable
import os
import sys

### initiate the centroids
def initiate_centroids(k, dset):
    '''
    Select k data points as centroids
    k: number of centroids
    dset: pandas dataframe
    '''
    centroids = dset.sample(k)
    return centroids



def plot_data(title, values, rng = None):

    c_pool = ['blue','green','red','cyan','magenta','yellow','orange','purple']

    Time = ['2m', '6m', '10m']
    N_Time = len(Time)
    PolyQ = ['Q80', 'Q92', 'Q111', 'Q140', 'Q175']
    N_PolyQ = len(PolyQ)
    Fourier = [0]
    #Fourier = [0,10,20,30]
    N_Fourier = len(Fourier)
    relTimeArr = np.array([0,4,8])
    relPolyQArr = np.array([0, 12, 31, 60, 95])
    tv = np.repeat(np.array([relTimeArr]),N_PolyQ,axis = 0)
    qv = np.repeat(np.transpose(np.array([relPolyQArr])),N_Time,axis = 1)

    FullGene = values
    currentGene = FullGene.reshape(N_PolyQ,N_Time)
        
    fig = plt.figure(figsize = (8,6.7))
    ax = fig.add_subplot(111, projection = '3d')
    ax.plot_wireframe(qv,tv,currentGene,label = "LFC", color = c_pool[0])
    lgd = ax.legend(bbox_to_anchor = (1.1, 1.1))
    lbly = ax.set_ylabel('\n'+'Age')
    ax.set_yticks(relTimeArr)
    ax.w_yaxis.set_ticklabels(Time)
    lblx = ax.set_xlabel('\n\n'+'Poly Q length')
    ax.set_xticks(relPolyQArr)
    ax.w_xaxis.set_ticklabels(PolyQ)
    ax.set_zlabel('LFC',rotation = 'vertical')
    ax.set_title(title,x = 0.5,y = 1.1)
    plt.setp(ax.get_xticklabels(),rotation = 30)
    if not rng is None:
        ax = plt.gca()
        ax.set_zlim(rng)
```

```python
### import samples

data_micro = pd.read_csv('/Users/thithanhyen/AResearch/Chritian_result/MiRAMINT_2.0/test_set/LFC_mir.txt', delimiter="\t")
data_mess = pd.read_csv('/Users/thithanhyen/AResearch/Chritian_result/MiRAMINT_2.0/test_set/LFC_gene.txt', delimiter="\t")

colnames = list(data_micro.columns)
```

```python
### set the parameter
X = data_mess.values
k_1 = 5

Y = data_micro.values
k_2 = 5
```

# 1. Analyse mRNA

## Classical k-means

```python
### k-mean apply to mRNA: denote by X.

kmeans = KMeans(n_clusters=k_1, random_state=0).fit(X)
#kmeans.labels_

#kmeans.predict([[0, 0], [12, 3]])
centroid_mess = kmeans.cluster_centers_
rng_min = np.min(centroid_mess)
rng_max = np.max(centroid_mess)
rng_mess = np.array([rng_min, rng_max])
for i in range(k_1):
    if i == 0:
        suff = 'st'
    elif i == 1:
        suff = 'nd'
    elif i == 2:
        suff = 'rd'
    else:
        suff = 'th'
    values = centroid_mess[i]
    title = str(i+1)+ suff + ' centroid'
    plot_data(title, values, rng_mess);
    plt.savefig(str(i+1)+ suff + '_centroid'+ '_mRNA')
```

```python
kmeans.labels_
```

```python
label_mess = np.array(list(kmeans.labels_))
```

```python
print('mRNA, the number of mRNAs in each cluster:\n')
for i in range(k_1): 
    if i == 0:
        suff = 'st'
    elif i == 1:
        suff = 'nd'
    elif i == 2:
        suff = 'rd'
    else:
        suff = 'th'
    nb_group = np.sum(label_mess == i)
    print(str(i+1)+ suff, ':', nb_group)
```

## Wasserstein k-means
### mRNA
Constrained k-means,(based on the article "Fast computations of Wasserstein Barycenters"

```python
# number of centroids
k = k_1

N, d = X.shape

measures_locations = [X,]
measures_weights = [np.ones(N)/N,]

#df = data_mess.copy()
#centroids = initiate_centroids(k, df)
#X_init = centroids.values
X_init = centroid_mess

b = np.ones((k,)) / k # initial Dirac locations


centroid_mess_W = ot.lp.free_support_barycenter(measures_locations, measures_weights, X_init, b)
```

```python
centroid_mess_W
```

```python
for i in range(k):
    values = centroid_mess_W[i]
    title = 'centroid_'+str(i)
    plot_data(title,values);
```

# 2. Analyse miRNA

## Classical k-means

```python
### k-mean apply to mRNA: denote by Y.
kmeans = KMeans(n_clusters=k_2, random_state=0).fit(Y)


#kmeans.predict([[0, 0], [12, 3]])
centroid_micro = kmeans.cluster_centers_
rng_min = np.min(centroid_micro)
rng_max = np.max(centroid_micro)

rng_micro = np.array([rng_min, rng_max])
for i in range(k_2):
    if i == 0:
        suff = 'st'
    elif i == 1:
        suff = 'nd'
    elif i == 2:
        suff = 'rd'
    else:
        suff = 'th'
    values = centroid_micro[i]
    title = str(i+1) + suff + ' centroid'
    plot_data(title, values, rng_micro);
    plt.savefig(str(i+1)+ suff + '_centroid'+ '_miRNA')
```

```python
label_micro = np.array(list(kmeans.labels_))
```

```python
print('miRNA, the number of miRNAs in each cluster:\n')
for i in range(k_2): 
    if i == 0:
        suff = 'st'
    elif i == 1:
        suff = 'nd'
    elif i == 2:
        suff = 'rd'
    else:
        suff = 'th'
    nb_group = np.sum(label_micro == i)
    print(str(i+1)+ suff, ':', nb_group)
```

## Wasserstein k-means

Constrained k-means,(based on the article "Fast computations of Wasserstein Barycenters"

```python

k = k_2

M, d = Y.shape

measures_locations = [Y,]
measures_weights = [np.ones(M)/M,]

#df = data_mess.copy()
#centroids = initiate_centroids(k, df)
#X_init = centroids.values
X_init = centroid_mess
# initial Dirac locations
b = np.ones((k,)) / k


centroid_micro_W = ot.lp.free_support_barycenter(measures_locations, measures_weights, X_init, b)
```

```python
centroid_micro_W
```

```python
for i in range(k):
    values = centroid_micro_W[i]
    title = 'centroid_'+str(i)
    plot_data(title,values);
```
